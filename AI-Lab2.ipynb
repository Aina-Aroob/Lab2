```python
import numpy as np
import matplotlib.pyplot as plt

file = open("genome.txt", "r")
# Read the first line
line = file.readline().strip() # strip newline characters from the line
genome_sequence=[]
for char in line:
  genome_sequence.append(char)

genome_length=len(genome_sequence)

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length) # z increases linearly to
color_map = {
    'A': 'red',
    'T': 'blue',
    'C': 'green',
    'G': 'yellow'
}

# Assign colors to each nucleotide in the genome sequence
colors = [color_map[base] for base in genome_sequence if base in color_map] # filter out invalid bases

# Plotting the 3D helix
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot each point on the helix
ax.scatter(x, y, z, c=colors, s= 100)  # s is the size of the points

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Helix of Genome Sequence')

# Show plot
plt.show()

# Combine the coordinates into a (genome_length x 3) array
coordinates = np.column_stack((x, y, z))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests



dora_img = Image.open("doraemon.jpg")

# Step 2: Convert the image to a numpy array
img_array = np.array(dora_img)

# Step 3: Plot the original image
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(img_array)
plt.title('Original Image')
plt.axis('off')

# Step 4: Rotate the image 90 degrees
rotated_img = np.rot90(img_array)

plt.subplot(2, 2, 2)
plt.imshow(rotated_img)
plt.title('Rotated Image')
plt.axis('off')

# Step 5: Flip the image left to right
flipped_img = np.fliplr(img_array)

plt.subplot(2, 2, 3)
plt.imshow(flipped_img)
plt.title('Flipped Image')
plt.axis('off')

# Step 6: Convert the image to grayscale
gray_img = np.dot(img_array[..., :3], [0.299, 0.587, 0.114])

plt.subplot(2, 2, 4)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Show all plots
plt.tight_layout()
plt.show()
#================================
x = img_array.mean(axis=1)
y = x * x

# plotting
plt.title("Line graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y, color ="blue")
plt.title('original image plot graph')
plt.show()

#================================
x = rotated_img.mean(axis=1)
y = x * x

# plotting
plt.title("Line graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y, color ="black")
plt.title('rotated image plot graph')
plt.show()

#=================================
x = flipped_img.mean(axis=1)
y = x * x

# plotting
plt.title("Line graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y, color ="purple")
plt.title('flipped image plot graph')
plt.show()

#=================================
x = gray_img.mean(axis=1)
y = x * x

# plotting
plt.title("Line graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y, color ="red")
plt.title('Grayscale plot graph')
plt.show()

```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
iris = load_iris()

# Accessing the features (data) using NumPy array
X = np.array(iris.data)
Y = np.array(iris.target)

# mean, median, and standard deviation for each feature
means = np.mean(X, axis=0)
medians = np.median(X, axis=0)
std_devs = np.std(X, axis=0)

# min and max values for each feature
mins = np.min(X, axis=0)
maxs = np.max(X, axis=0)

# Extract only the sepal length and sepal width as a NumPy array
sepal_data = X[:, :2]  # Only sepal length and sepal width

# Print results
print("Means of features:", means)
print("Medians of features:", medians)
print("Standard Deviations of features:", std_devs)
print("Minimum values of features:", mins)
print("Maximum values of features:", maxs)
#print("Sepal Length and Width Data:\n", sepal_data)

# Step 2a: Create a scatter plot of sepal length vs sepal width
plt.figure(figsize=(15, 5))

# Scatter plot for Sepal Length vs Sepal Width
plt.subplot(1, 3, 1)
plt.scatter(sepal_data[:, 0], sepal_data[:, 1], c=Y, cmap='viridis', edgecolor='k', s=100)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Step 2b: Plot a histogram showing the distribution of sepal length
plt.subplot(1, 3, 2)
plt.hist(X[:, 0], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

# Step 2c: Create a line plot to visualize the relationship between petal length and petal width
plt.subplot(1, 3, 3)
plt.plot(X[:, 2], X[:, 3], 'o-', color='orange')  # Petal Length vs Petal Width
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

# Show all plots
plt.tight_layout()
plt.show()
```
